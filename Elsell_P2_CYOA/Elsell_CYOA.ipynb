```python
from colorama import Fore as f, Back as b, Style as s
import sys
import random
```

```python
global uInput
global hp
global pN
global t
uInput = 0


def randColors(instr = "test"): #A function that returns the input string in random terminal colors
    
    #term color values
    W  = '\033[0m' 
    R  = '\033[31m' 
    G  = '\033[32m' 
    O  = '\033[33m'
    B  = '\033[34m'
    P  = '\033[35m'
    BL = '\033[36m'
    
    colors = [W, R, G, O, B, P, BL]
    ostr = ""
    for i in instr:
        ostr += i + random.choice(colors)
    return ostr

def genMem(): # Genrate random number of cult mumbers 
    randCult = random.randint(2,7) #Genearate a random 
    return randCult


def genSRate(n, a, h): 
    if(n <= 5 and a == True and h > 5): #they got the pistol and there helath is < then 5 and the number of cult members is <= five
        randSrate = round(random.uniform(0.6, 1), 2) #generate a survival rate between 60 and 100 percent
        #print("High Rate" + str(randSrate))
    else:
        randSrate = round(random.uniform(0, 0.51), 2) #generate a survival rate between 0 and 50 perecent
    
    return randSrate


def select(): #user input function
    a = int(input(f.MAGENTA+"Selection: "))
    return a
    

def printStats(t,h): #print out the current player stats to screen
    print(b.BLUE+"Turn: " + str(t) +"\nHealth: " + str(h)+b.RESET)
    

def resInput(): #Reset the user input to 0 used in edge cases
    uInput = 0
    return uInput

def gOver(hp, ec): # check if the game is over with 3 cases 
    gameO = open('gover.txt', 'r')
    gameOcont = gameO.read()
    
    gO ="Game Over!"
    eT = "Ending Type: "
    if(hp < 1):#and
        pBreak()
        print(b.BLACK+s.BRIGHT+f.RED+gameOcont) # Print gamer over ascii
        gameO.close()
        print(b.RED+ eT + ec)
        sys.exit(0)
    elif(ec == "It was all a dream?"): #for the combo choice 2,2 (this choice decreases hp but not enough to kill the player)
        pBreak()
        mth = "MultiVerse-Theory-MultiVerse-Theory"
        mthLen = len(mth)
        print(b.BLACK+s.BRIGHT+f.MAGENTA+gameOcont) # Print gamer over ascii
        gameO.close()
        print(b.RED+eT+s.RESET_ALL)
        j=1
        while j < mthLen:
            print(randColors(mth))
            j +=1
            if j > mthLen:
                break;
        sys.exit(0)
        
    elif(ec == "The good one"): #Special game over for the best ending
        print(f.GREEN+"=======================================================")
        print(randColors("======================================================="))
        print(randColors("======================================================="))
        print(randColors("======================================================="))
        print(randColors("=======================================================")+f.RESET)
        print(b.BLUE+s.BRIGHT+"The game is over and you got the best ending possible.\nYour choices and a bit of luck got you here.\nThank you for playing"+b.RESET)
        print(randColors("======================================================="))
        print(randColors("======================================================="))
        print(randColors("======================================================="))
        print(randColors("======================================================="))
        print(f.GREEN+"=======================================================")
        sys.exit(0)


        
def eTurn(h, t, ec): #End the turn by printing the stats and checkign g over
    t= t +1
    newG.turn = t #update the value of turn in newG and set it equal to the value passed in + 1
    printStats(t,h) 
    gOver(h, ec) # check if gameover
    turn = t;
    
    return h and t # return these values back to the  function which called eTurn

def pBreak(): #PageBreak
    print(f.BLACK+"============================================================================")

def hlp(): #print help menu
    pBreak()
    print(s.BRIGHT+f.RED+"////////////////////////////Help Menu\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\\")
    print("This is text based RPG.")
    print("Use your keyboard and use the corresponding numbers to make choices.")
    print("The choices you make directly effect the probability of your success.")
    print(b.YELLOW+"If you enter a invalid choice while playing you will have to restart the game!"+s.RESET_ALL)
    pBreak()
    mainMenu()
    
def quit(): # quit the game
    pBreak()
    print(f.CYAN+"Quitting")
    pBreak()
    sys.exit(0)
    
def mainMenu(): 
    pBreak()
    ts = open('title.txt', 'r')
    tsC = ts.read()
    print(b.BLACK+s.BRIGHT+f.YELLOW+tsC+s.RESET_ALL) #print title ascii
    ts.close()
          
    print(b.BLACK+f.YELLOW +s.BRIGHT+"Welcome to Crypt Crawler:")
    print("Use your keyboard to make a selection:")
    print("1. New Game\n2. Help\n3. Quit"+s.RESET_ALL)
    uInput = select()
    pBreak()
    if(uInput == 1):
        newG()
    elif(uInput == 2):
        hlp()
    elif(uInput == 3):
        quit()
    else:
        hlp()

def newG(): # New Game is basically the function that bulk of the content it controls the structure along with calliing alot of the helper functions
    hp = 5 #Health
    ec = "null" #end Code
    turn = 1
    cMem = 0 # Number of cult memebers
    sRate = 0 # Survival Rate
    armed = False #Did the player get the gun?
    
    pBreak()
    pN = input(f.BLUE + "What's your name adventurer: ")
    print(f.GREEN+"You emerge from a deep sleep-in mid-day.\nYou find yourself near a entrance to cave unable to recall a thing.")
    print(f.BLUE+"Do you (1) enter the cave or (2) remain at the cave or remain at the entrance?"+f.RESET)
    uInput =select()
    eTurn(hp, turn, ec)
    if(uInput == 1): #enter cave
        pBreak()
        print(f.GREEN+"You enter the cave and as you venture deeper and deeper your drawn to a light:")
        print(f.BLUE+"Do you (1) investigate the light or (2) venture deeper?"+f.RESET)
        turn += 1
        eTurn(hp, turn, ec)
        uInput = select()
        if(uInput == 1): #chase the light
            print(f.GREEN+"You enter a chasm with high vaulted ceilings as you approach the light source a booming voice echoes:")
            print(f.GREEN+"Hello " + pN +" we've been expecting you!")
            print("Suprised and in fear you grab the light source which turned out be a torch on the wall.\nWithout warning the entrance from which you came collapses and you notice the ceiling will soon follow.\nYou sprint further into the depths torch in hand and come to a fork.")
            print(f.BLUE+"Do you go (1) left or (2) right?"+f.RESET)
            #turn +=1
            eTurn(hp, turn, ec)
            uInput = select()
            if(uInput == 1): # go left
                print(f.GREEN+"You take a left at the fork and after a few minutes of walking you are extremely thirsty.\nYou come upon a clean looking resivor of water.")
                print(f.BLUE+"Do you (1) stop to drink or (2) continue walking?"+f.RESET)
                turn +=1
                eTurn(hp, turn, ec)
                uInput = select()
                if(uInput == 1): # drink water (hp buff)
                    hp +=1
                    print(f.GREEN+"You stop to drink and find the water refreshing pausing to regain strength.")
                    print(f.GREEN+"You get up from resting an proceed down the cave"+f.RESET)
                    turn+=1
                    eTurn(hp, turn, ec)
                    print(f.GREEN+"You here voices in the distance.")
                    print(f.BLUE+"Do you (1) investage or (2) hide"+f.RESET)
                    uInput = select()
                    if(uInput == 1): #cult kills w/ hp buff
                        print(f.GREEN+"You follow the voices and come upon a room of figures in red cloaks.\nAs you enter you kick a pebble alerting them of your presence.\nThey murder you in cold blood"+f.RESET)
                        hp-=6
                        ec = "Murdered by a cult"
                        turn+=1
                        eTurn(hp, turn, ec) 
                    elif(uInput == 2): #cult doesn't kill w/ hp buff
                        cMem = genMem()
                        print(f.GREEN+str(cMem)+" hooded figures pass you as you hide behind a stalgmite."+f.RESET)
                        eTurn(hp, turn, ec) 
                        print(f.GREEN+"You proceed to the room from which they came and find a chest.")
                        print(f.BLUE+"Do you (1) open or (2) leave it closed?"+f.RESET)
                        turn+=1
                        eTurn(hp, turn, ec)
                        uInput = select()
                        if(uInput == 1):
                            print(f.GREEN+"You find a pistol with ammo and map of the cave.\n Along with pictures of yourself as a child with the words " + f.RED+"SATONY ZAZAS"+f.RESET)
                            print(f.GREEN+"You hear voices in the disatance prepare for a fight."+f.RESET)
                            armed = True
                            sRate = genSRate(cMem, armed, hp)
                            if (sRate >= 0.5):
                                print(f.RED+"You see figures enter the room and proceed to attack a second cult member grabs you throws you to the ground.\nYou push them off you and fire your weapon killing one of the two.\nYou regain your footing a fire a shot into the remaining cult member.")
                                print(f.GREEN+"It's over you use the map to navigate out of the cave and barely espace"+f.RESET)
                                ec = "The good one"
                                turn+=1
                                eTurn(hp,turn,ec)
                            elif(sRate < 0.5):
                                print(f.RED+"You see figures enter the room and proceed to attack a second cult member grabs you throws you to the ground.\nThe first cult member says 'Goodbye " + pN +"' proceeding to step on your face and everything goes"+b.WHITE+" black."+f.RESET+b.RESET)
                                ec = "The bad one"
                                hp-=10;
                                turn+=1
                                eTurn(hp,turn,ec)
                            
                        elif(uInput == 2):
                            print(f.GREEN+"You leave the chest closed and begin to look around the room and see the words "+f.RED+"SATONY ZAZAS"+f.GREEN+" everywhere along with pictures of you at various ages."+f.RESET)
                            print(f.GREEN+"You hear voices in the disatance prepare for a fight.")
                            if (sRate >= 0.5):
                                print(f.RED+"You see figures enter the room and proceed to attack a second cult member grabs you throws you to the ground dropping a pistol in the process.\nYou push them off you and fire the weapon killing one of the two.\nYou regain your footing a fire a shot into the remaining cult member.")
                                print(f.GREEN+"It's over you get up and find a map on the ground using it to navigate out of the cave and barely espace")
                                ec = "The good one"
                                turn +=1
                                eTurn(hp,turn,ec)
                            elif(sRate < 0.5):
                                print(f.RED+"You see figures enter the room and proceed to attack a second cult member grabs you throws you to the ground.\nThe first cult member says 'Goodbye " + pN +"' proceeding to step on your face and everything goes"+b.WHITE+" black."+f.RESET+b.RESET)
                                ec = "The bad one"
                                hp-=10;
                                turn +=1
                                eTurn(hp,turn,ec)
                    
                elif(uInput ==2): # don't drink the water (hp hit)
                    hp -= 1
                    print(f.GREEN+"You continue on your path but slowly begin to feel increasingly thirsty and weak."+f.RESET)
                    eTurn(hp, turn, ec)
                    print(f.GREEN+"You here voices in the distance.")
                    print(f.BLUE+"Do you (1) investage or (2) hide"+f.RESET)
                    uInput = select()
                    if(uInput == 1): #cult kills with w/ hp decrease
                        print(f.GREEN+"You follow the voices and come upon a room of figures in red cloaks.\nAs you enter you kick a pebble alerting them of your presence.\nThey murder you in cold blood"+f.RESET)
                        hp-=5
                        ec = "Murdered by a cult"
                        turn +=1
                        eTurn(hp, turn, ec) 
                    elif(uInput == 2): #cult doesnt kill w/ hp decrease
                        cMem = genMem()
                        print(f.GREEN+str(cMem)+" hooded figures pass you as you hide behind a stalgmite."+f.RESET)
                        eTurn(hp, turn, ec)
                        print(f.GREEN+"You proceed to the room from which they came and find a chest.")
                        print(f.BLUE+"Do you (1) open or (2) leave it closed?"+f.RESET)
                        turn +=1
                        eTurn(hp, turn, ec)
                        uInput = select()
                        if(uInput == 1):
                            print(f.GREEN+"You find a pistol with ammo and map of the cave.\n Along with pictures of yourself as a child with the words " + f.RED+"SATONY ZAZAS"+f.RESET)
                            print(f.GREEN+"You hear voices in the disatance prepare for a fight."+f.RESET)
                            armed = True
                            sRate = genSRate(cMem, armed, hp)
                            if (sRate >= 0.5):
                                print(f.RED+"You see figures enter the room and proceed to attack a second cult member grabs you throws you to the ground.\nYou push them off you and fire your weapon killing one of the two.\nYou regain your footing a fire a shot into the remaining cult member.")
                                print(f.GREEN+"It's over you use the map to navigate out of the cave and barely espace"+f.RESET)
                                ec = "The good one"
                                turn +=1
                                eTurn(hp,turn,ec)
                            elif(sRate < 0.5):
                                print(f.RED+"You see figures enter the room and proceed to attack a second cult member grabs you throws you to the ground.\nThe first cult member says 'Goodbye " + pN +"' proceeding to step on your face and everything goes"+b.WHITE+" black."+f.RESET+b.RESET)
                                ec = "The bad one"
                                hp-=10;
                                turn +=1
                                eTurn(hp,turn,ec)
                            
                        elif(uInput == 2):
                            print(f.GREEN+"You leave the chest closed and begin to look around the room and see the words "+f.RED+"SATONY ZAZAS"+f.GREEN+" everywhere along with pictures of you at various ages."+f.RESET)
                            print(f.GREEN+"You hear voices in the disatance prepare for a fight.")
                            if (sRate >= 0.5):
                                print(f.RED+"You see figures enter the room and proceed to attack a second cult member grabs you throws you to the ground dropping a pistol in the process.\nYou push them off you and fire the weapon killing one of the two.\nYou regain your footing a fire a shot into the remaining cult member.")
                                print(f.GREEN+"It's over you get up and find a map on the ground using it to navigate out of the cave and barely espace")
                                ec = "The good one"
                                turn +=1
                                eTurn(hp,turn,ec)
                            elif(sRate < 0.5):
                                print(f.RED+"You see figures enter the room and proceed to attack a second cult member grabs you throws you to the ground.\nThe first cult member says 'Goodbye " + pN +"' proceeding to step on your face and everything goes"+b.WHITE+"black."+f.RESET+b.RESET)
                                ec = "The bad one"
                                turn +=1
                                hp-=10;
                                eTurn(hp,turn,ec)
                        
                    
            elif(uInput == 2): #You go right
                print(f.GREEN+"You take a right at the fork and walk for half an hour.\nYou come upon the ruins of a mine shaft.\nAs you continue to walk you notice rocks falling from the ceiling and the shaft is increasingly more unstable.\nA large boulder falls from the shaft and pins your leg against the cold ground of the cave.\nYou are too weak to lift it."+f.RESET)
                hp-=5
                ec = "stuck under a rock"
                turn +=1
                eTurn(hp, turn, ec)
            
        elif(uInput == 2): #You get lost without a torch End
            print(f.GREEN+"You venture deeper into the chasm and as it begins to get darker you become disoriented.\nYou hear voice in the distance and try to approach leading you even further.\nYou eventually collapse from mal nourishment and accept your death"+f.RESET)
            hp-=5
            ec = "Lost & Starving"
            turn +=1
            eTurn(hp, turn, ec)
           
    elif(uInput == 2): #wait to enter cave
        print(f.GREEN+"You build a fire and begin to relax several hours pass.\nYou fall back asleep and are abruptly awoken by a loud rustling.\n")
        print(f.BLUE+"Do you (1) you fight the bear or (2) run to the cave and take shelter?"+f.RESET)
        uInput = select()
        eTurn(hp, turn, ec)
        if(uInput == 1): #fight bear End
            print(f.GREEN+"You immediately get struck to the ground as the adrenaline courses through your veins everything goes black."+f.RESET)
            hp -=5
            turn +=1
            ec = "Mauled By a Bear"
            eTurn(hp, turn, ec)
        elif(uInput == 2): #You run from the bear End
            print(f.GREEN+"You swiftly sprint to the cave but trip over a nearby log cutting your leg in the process."+f.RESET)
            hp -=1
            eTurn(hp, turn, ec)
            print(f.GREEN+"As you enter the cave you here a beeping the further you go the louder it gets.\nYou see a light source in the distance and approach.\nOnce you reach it your vision is consumed by blinding white.\nYou wake up disoriented in your bed and exclaim 'what a strange dream'!\nYou look down at your leg and see a large cut.\nConfused you start to question if it was a dream."+f.RESET)
            ec = "It was all a dream?"
            turn +=1
            eTurn(hp, turn, ec)
            
            
```
