```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
%matplotlib inline

plt.rcParams["figure.figsize"] = (8, 6)
```

```python
superfemale = pd.read_csv("~/Documents/INFO1201/Mod 5 Visualizing Data/superhero_data_copy.csv", parse_dates = ["FIRST_YEAR"])
```

#### **Question 1** 
##### *What is the alignment of the new characters introduced by year?*
For the opening question, I first wanted to know how many bad and good characters were introduced over the years and to see the trend of each alignment. I started with the .valuecounts() to see how many alignment characters were there. Based on data from the output, I decided to remove ND (No Data) and Reformed Criminals from the dataset. This decision was made because I only wanted to know about bad and good criminals and the numbers for Reformed criminals were very low and almost non-existent when graphed. I included neutral characters because I was surprised to see the number of those characters being introduced to be more than what I expected. The removal of ND and Reformed Criminals data did not affect my graph significantly. 

I used pd.crosstab to analyze the data and used the .replace and .drop function to clean the data frame. I chose a line graph because it was the best way to see the data across time and also visually represented the numbers for the data. 

```python
years= superfemale

years= years.replace("ND", "Delete")
#years = years.replace("Neutral Characters", "Delete")
years = years.replace("Reformed Criminals", "Delete")
years_new= years.drop(years[years.ALIGN== "Delete"].index)
years_new= years_new.replace("ND", "Delete")
old_years= years_new.drop(years_new[years_new.FIRST_YEAR== "Delete"].index)

prettyyears= pd.crosstab(old_years['FIRST_YEAR'], old_years['ALIGN'])
prettyyears.plot(color=([ "darkolivegreen", "palegreen", "darkseagreen",]))
plt.title("Number of New Characters by Year", fontsize= 30, loc= "center", color= "darkgreen")
plt.xlabel("Years", fontsize = 14)
plt.ylabel("Number of Characters", fontsize = 14)
plt.show()
```

##### *Analysis*
The overall trend that I noticed almost immediately was that bad characters were introduced more than good characters. In the graph, you can see that the most amount of bad characters introduced was between 1986-1996. The good characters did follow the same trend as bad characters; however, it was still not as much as the bad characters. It about about 100 characters less than bad characters. 

#### **Question 2**
##### *What is the alignment of characters in each universe?*
After seeing the trend through the years, I now wanted to know which universe was creating the most good and bad characters. For this data frame, I included reformed criminals and neutral characters. Even though the reformed criminals bar is not seen, there was data present and I thought that it was still important to include. The only data that was omitted was ND because it did not provide anything useful for the purpose of this graph. 

I used pd.crosstab to analyze the data and used the .replace and .drop function to clean the data frame. I chose a bar graph because it was the best way to see the difference between the two genders and see the variety within the multiple bars. 

```python
align= superfemale.drop(superfemale[superfemale.ALIGN== "ND"].index)
#date= align["ALIGN"].value_counts()
#print(date)

evil_characters= pd.crosstab(align["ALIGN"], align["UNIVERSE"])
evil_characters.plot(kind = 'barh', color=(["navy", "powderblue"]))
plt.title("Aligment of Characters by Universe", fontsize = 30, loc= "center", color= "midnightblue")
plt.xticks(rotation = 0, fontsize = 10)
plt.yticks(fontsize = 12)
plt.xlabel("Number of Characters", fontsize = 14)
plt.ylabel("Alignment", fontsize = 14)
plt.show()
```

##### *Analysis*
I was very surprised with this result. Marvel introdcued the most bad characters and good characters. They introduced 2000 more bad characters than good characters. They also introduced more neutral characters than DC. DC introduced bad and good characters equally with bad characters slightly more. 

#### **Question 3**
##### *How many male and female characters have a secret identity versus a public identity?*
Shifting gears, I wanted to explore if gender played a factor into the identity status of characters. For this particular data frame, I had to omitted ND, Agender, Genderfluid and Genderless characters because I wanted to focus solely on gender binary characters. With identity status, I added Nathaniel Richards into the secret identity category after completing a simple search for that character. The ND for the ID dataframe was also removed. 

I used pd.crosstab to analyze the data and used the .replace and .drop function to clean the data frame. I chose a bar graph because it was the best way to see the difference between the two genders and see the variety within the multiple bars. 

```python
#female= superfemale["SEX"].value_counts()
#print(female)
change_gender = superfemale
change_gender= change_gender.replace("Male Characters", "Male")
change_gender= change_gender.replace("Male Character", "Male")
change_gender= change_gender.replace("Female Characters", "Female")
change_gender= change_gender.replace("Female Character", "Female")
change_gender= change_gender.replace("ND", "Delete")
change_gender= change_gender.replace("Agender Characters", "Delete")
change_gender= change_gender.replace("Genderfluid Characters", "Delete")
change_gender= change_gender.replace("Genderless Characters", "Delete")
new_gender= change_gender.drop(change_gender[change_gender.SEX == "Delete"].index)
new_gender= new_gender.replace("Nathaniel Richards", "Secret Identity")
newi_gender= new_gender.drop(new_gender[new_gender.ID== "Delete"].index)


male= pd.crosstab(newi_gender["SEX"], newi_gender["ID"])
male.plot(kind="bar", fontsize=14, color=(["thistle","rebeccapurple", "plum","violet", "purple"]))
plt.title("Gender and Identity Status", fontsize = 30, loc= "center", color= "darkmagenta")
plt.ylabel("Number of Characters", fontsize = 14)
plt.show()
```

##### *Analysis*
In general, there is obviously more male characters than female characters due to the gender disparity in the comic book industry. There was 4x (four times) more male characters than female characters that have secret identities. For male characters, there are more characters with secret identity than with public identity. There also very few characters with no dual identity and identities known to the authorities. 

#### **Question 4**
##### *How many bad characters have a secret identity as compared to good characters by gender?*
With this question, I wanted to combine previous data exploration and visually analyze the data around identity status, alignment and gender. For this particular data frame, I had to omitted ND, Agender, Genderfluid and Genderless characters because I wanted to focus solely on gender binary characters. With alignment dataframe, I removed ND, Neutral characters and Reformed criminals to soley focus on good and bad characters. 

I used pd.crosstab to analyze the data and used the .replace and .drop function to clean the data frame. I chose a stacked bar graph because it best represented the crosstab data of multiple variables and one can visually see the differences in data.

```python
updated_superfemale = superfemale

updated_superfemale = updated_superfemale.replace("Nathaniel Richards", "Secret Identity")
updated_superfemale = updated_superfemale.replace("ND", "Delete")
update = updated_superfemale.drop(updated_superfemale[updated_superfemale.ID == "Delete"].index)

update= update.replace("Male Characters", "Male")
update= update.replace("Male Character", "Male")
update= update.replace("Female Characters", "Female")
update= update.replace("Female Character", "Female")
update= update.replace("ND", "Delete")
update= update.replace("Agender Characters", "Delete")
update= update.replace("Genderfluid Characters", "Delete")
update= update.replace("Genderless Characters", "Delete")
update_again= update.drop(update[update.SEX== "Delete"].index)

update_again= update_again.replace("ND", "Delete")
update_again = update_again.replace("Neutral Characters", "Delete")
update_again = update_again.replace("Reformed Criminals", "Delete")
update_onceagain= update_again.drop(update_again[update_again.ALIGN== "Delete"].index)


evil_understood= pd.crosstab(index= [update_onceagain["SEX"], update_onceagain["ALIGN"]], columns = [update_onceagain["ID"]])
evil_understood.plot(kind = 'bar', stacked = True, color=(["darkslategrey", "lightcyan", "paleturquoise","mediumturquoise", "teal"]))

plt.title("The Gender, Alignment and their Identity Status", fontsize = 30, loc= "center", color= "darkcyan")
plt.xticks(rotation=-10, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel("Gender and Alignment", fontsize = 16)
plt.ylabel("Number of Characters", fontsize = 16)
plt.show()
```

##### *Analysis*
Based on this graph, male and bad characters were had the most amount of characters as we observed before. In this category, there are more characters with a secret identity. For female characters, there are more bad characters whose identities are secret than public compared to good characters. 

#### **Question 5**
##### *How many bad characters are dead or alive by gender?*
For the final question, I wanted to know if bad characters were often killed by the good characters. For this particular data frame, I had to omitted ND, Agender, Genderfluid and Genderless characters because I wanted to focus solely on gender binary characters. With alignment, I also omitted ND data points. 

I used pd.crosstab to analyze the data and used the .replace and .drop function to clean the data frame. I chose a stacked bar graph because it best represented the crosstab data of multiple variables and one can visually see the differences in data.

```python
alive= superfemale
alive= alive.replace ("ND", "Delete")
alive= alive.replace ("Male Character", "Male")
alive= alive.replace ("Male Characters", "Male")
alive= alive.replace ("Female Characters", "Female")
alive= alive.replace ("Agender Characters", "Delete")
alive= alive.replace ("Genderfluid Characters", "Delete")
alive= alive.replace ("Transgender Characters", "Delete")
alive= alive.replace ("Genderless Characters", "Delete")
alive_one= alive.drop(alive[alive.SEX=="Delete"].index)
alive_one= alive_one.replace("ND", "Delete")
alive_two= alive_one.drop(alive_one[alive_one.ALIVE== "Delete"].index)
alive_two= alive_two.replace ("ND", "Delete")
alive_three= alive_two.drop(alive_two[alive_two.ALIGN== "Delete"].index)
aliveordead= pd.crosstab(index= [alive_three["ALIGN"], alive_three["SEX"]], columns= alive_three["ALIVE"])

aliveordead.plot(kind = 'bar', stacked = True, color=(["indigo","thistle"]))

plt.title("Living Characters With Their Gender And Alignment", fontsize = 30, loc= "center", color= "rebeccapurple")
plt.xticks(rotation=-90, fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel("Gender and Alignment", fontsize = 16)
plt.ylabel("Number of Characters", fontsize = 16)
         
plt.show()
```

##### *Analysis*
I was surprised by this result as I thought there would be more bad characters dead. However, there were more bad male characters who were alive. The most amount of deceased characters is under bad male characters. With female characters, there is an equal amount of dead and alive characters.
