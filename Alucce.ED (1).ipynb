```python
#import the libraries we want to use
import pandas as pd
import numpy as np
```

```python
#load up our csv file into python
superheros = pd.read_csv("superhero_data.csv" , parse_dates = ["FIRST_YEAR"])
```

```python
superheros.head()
```

```python
#introduction of the dataframe. From Pandas (pd) library
df = pd.DataFrame(data = superheros)
print(df)
```

```python
#What are different type of hair the superheros has?
pd.crosstab(superheros["HAIR"], superheros["ALIGN"])
```

```python
#How many haroes have differnt type of hair and eyes?
bad_characters = pd.crosstab(index = superheros["HAIR"] == "Black Hair", columns = superheros["EYES"] == "Blue Eyes")

bad_characters.columns = ["Has blue eyes", "Does not have blue eyes"]
bad_characters.index = ["Has black hair", "Does not have black hair"]
bad_characters 
```

```python
#What are the different types of sexualityâ€™s of the characters? 
pd.crosstab(superheros["SEXUALITY"], superheros["ALIGN"], margins=True)
```
