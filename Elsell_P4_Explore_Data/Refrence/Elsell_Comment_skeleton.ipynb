```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = (8, 6)
#import dependencies, pandas, numerical python, matplotlib
```

```python
superheroes = pd.read_csv("superhero_data.csv", parse_dates = ["FIRST_YEAR"]) 
#sort the data set in chronological order by charcter first year of appearence
```

```python
superheroes.head() 
#confirm data IO (input/output)
```

## Plot #1: What suggestions would you make to describe this code? 

```python
appearance_by_sex = pd.crosstab(superheroes['FIRST_YEAR'], superheroes['ALIGN']) 
#compare the year vs the type of alignment and put into a crosstab matrix

appearance_by_sex.plot()
#plot the cross tab with def args
plt.title("Number of New Characters by Alignment and by Year", fontsize = 18) #Set title args
plt.xlabel("Years", fontsize = 14)#set x axis title
plt.ylabel("Number of Characters", fontsize = 14)#set y axis title
#above lines specify args for the graph in general.

plt.show() #show graph with args
```

Notice the ratios of bad vs good charcters at the points 1995 and 1945. The correlation here is that historically "bad charcters" outweighed "good charcters". Therefore superheroes must have been more powerful since generally this genre doesn't end in tradgedy. Mre supervillains where needed in the universe to make the plot interesting and create the illusion that the heroes could lose.

## Plot #2: How do I explain a more complicated, highly filtered data analysis with customized pie chart?

```python
genderA = superheroes.drop(superheroes[superheroes.SEX == "Agender Characters"].index) 
#remove a gender chars
genderGL = genderA.drop(genderA[genderA.SEX == "Genderless Characters"].index)
#remove gender less chars
genderGF = genderGL.drop(genderGL[genderGL.SEX == "Genderfluid Characters"].index)
genderT = genderGF.drop(genderGF[genderGF.SEX == "Transgender Characters"].index)
#The above lines clean the data and drop irrelevant infromation
genderMF = genderT

sex_counts = genderMF["SEX"].value_counts() 
# after dropping all the data sum the values of each object in the column.

sex_counts.plot(kind = "pie", autopct = "%1.0f%%", subplots = True, fontsize = 14) #plot values
#specify the plot type as pie, autopct for precentages, subplots, 
plt.title("Number of Marvel and DC Characters by Sex", fontsize = 18) #label title
plt.axis("equal")

plt.show()
```

## Plot #3: This crosstab-based plot is looking at 3 different dimensions: character count, alignment and identity. What comments would you add to help me understand the analysis and the plot is generates?  

```python
identity_count = pd.crosstab(superheroes['ALIGN'], superheroes['ID'])

identity_count.plot(kind= 'bar', stacked = True, width = 0.75)
plt.title('Number of Characters by Alignment and Identity', fontsize = 18)
plt.xlabel("Alignment Type", fontsize = 14)
plt.ylabel("Number of Characters", fontsize = 14)
plt.xticks(rotation = 20, fontsize = 10)
plt.yticks(fontsize = 12)

plt.show()
```

This bar chart shows the summation of Bad vs Good charcters against the status of their identities. This plot can answer correlations between in general do superheroes or supervillians perfer to be know to the public.
