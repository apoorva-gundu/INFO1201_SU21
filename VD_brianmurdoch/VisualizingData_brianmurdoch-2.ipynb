# 1. What percentage of male characters are dead?
  Compare all male characters to characters that are both male and deceased 

```python


# comparison of male characters to deceased males
slicesmale_slicesdead = [16420, 4140]
labels = ['Male Characters 16420', 'Dead Male Characters 4140']

#plot
plt.pie(slicesmale_slicesdead, labels=labels, wedgeprops={'edgecolor': 'black'}, autopct = '%2.1f%%')




plt.title("Ratio of male characters to deceased male characters", fontsize = 18)
plt.axis("equal")
plt.show()

```

 # 2. What percentage of female characters first appeared in april?
  Compare all female characters to female characters that also first apppeared in april 

```python


# comparison of all female characters to female characters that also first appeared in april
slicesfemale_slicesapril = [5803, 186]
labels = ['Female Characters 5803', 'Female Characters That First Appeared in April 186']

# plot 
plt.pie(slicesfemale_slicesapril, labels=labels, wedgeprops={'edgecolor': 'black'}, autopct = '%2.1f%%')




plt.title("Ratio of Female Characters to Female Characters that First Appeared in April", fontsize = 18)
plt.axis("equal")

plt.show()
```

# 3. What is the percentage of characters that have blue eyes and appeared more than 88 times?
  Compare the number of characters with blue eyes who have appeared more than 88 times to all characters
